## Calculation of the age of the patients

```python

def timedelta_to_years(tdelta):
    average_year_in_days = 365.25
    years = round(tdelta.total_seconds() / (average_year_in_days * 24 * 60 * 60))
    return years
```

```python
df = df.rename(columns={"timestamp_start_Unnamed: 2094_level_1": "timestamp_start"})
age_in_days = pd.to_datetime(df["timestamp_start"], format="mixed") - pd.to_datetime(df["birthday"], format="mixed")
df["age"] = age_in_days.apply(timedelta_to_years)
```
